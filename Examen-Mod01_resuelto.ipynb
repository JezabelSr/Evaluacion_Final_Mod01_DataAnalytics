```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []      
        self.ventas_totales = 0

    def agregar_producto (self, nombre, precio, cantidad):
        producto = {
        'nombre': nombre,
        'precio': precio,
        'cantidad': cantidad
        }
        for elemento in self.inventario:
            if elemento['nombre'] == nombre:
                elemento ['cantidad'] += cantidad
                break

        else:
            self.inventario.append(producto)

    def ver_inventario(self):
        for elemento in self.inventario: 
            print(f"Nombre: {elemento['nombre']}, Precio: ${elemento ['precio']}, Cantidad: {elemento ['cantidad']}")

    def buscar_producto (self, nombre):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre: 
                print(f"Nombre: {elemento['nombre']}, Precio: ${elemento['precio']}, Cantidad: {elemento['cantidad']}")
                break
        else:
            print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def actualizar_stock (self, nombre, cantidad):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre:
                elemento ['cantidad'] = cantidad
                print (f"Valor actualizado. Ahora hay {cantidad} unidades de {nombre} ")
                break
        else:
                print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def eliminar_producto (self, nombre):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre:
                self.inventario.remove(elemento)
                print (f"Ok, se han borrado {nombre} de tu inventario")
                break
        else:
                print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def calcular_valor_inventario(self):
        total = 0
        for elemento in self.inventario:
                total += elemento['precio'] * elemento['cantidad']

        print (f"El valor total del inventario es ${total}")


```
