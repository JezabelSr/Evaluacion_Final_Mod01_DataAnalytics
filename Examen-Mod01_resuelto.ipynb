```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []      
        self.ventas_totales = 0
        self.clientes = {}

    def agregar_producto (self, nombre, precio, cantidad):
        producto = {
        'nombre': nombre,
        'precio': precio,
        'cantidad': cantidad
        }
        for elemento in self.inventario:
            if elemento['nombre'] == nombre:
                elemento ['cantidad'] += cantidad
                break

        else:
            self.inventario.append(producto)

    def ver_inventario(self):
        for elemento in self.inventario: 
            print(f"Nombre: {elemento['nombre']}, Precio: ${elemento ['precio']}, Cantidad: {elemento ['cantidad']}")

    def buscar_producto (self, nombre):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre: 
                print(f"Nombre: {elemento['nombre']}, Precio: ${elemento['precio']}, Cantidad: {elemento['cantidad']}")
                break
        else:
            print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def actualizar_stock (self, nombre, cantidad):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre:
                elemento ['cantidad'] = cantidad
                print (f"Valor actualizado. Ahora hay {cantidad} unidades de {nombre} ")
                break
        else:
                print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def eliminar_producto (self, nombre):
        for elemento in self.inventario: 
            if elemento['nombre'] == nombre:
                self.inventario.remove(elemento)
                print (f"Ok, se han borrado {nombre} de tu inventario")
                break
        else:
                print (f"Vaya, el producto {nombre} no se ha encontrado en nuestra Base de Datos")

    def calcular_valor_inventario(self):
        total = 0
        for elemento in self.inventario:
                total += elemento['precio'] * elemento['cantidad']

        print (f"El valor total del inventario es ${total}")

    def agregar_cliente ( self, nombre, email ):
        self.clientes [nombre] = {'email': email, 'compras': []}

    def ver_clientes(self):
        for clave, valor in self.clientes.items():
            print(f"Nombre: {clave}, Email: {valor ['email']}")

    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente not in self.clientes: 
            print("El cliente no existe")

        else:
            total = 0
            for producto, detalles in carrito.items():
                total += detalles['precio'] * detalles['cantidad']
            self.clientes[nombre_cliente]['compras'].append({'productos': carrito, 'total': total})
            self.ventas_totales  += total
            print (f"Compra registrada correctamente para {nombre_cliente}")
            print(f"El total de tu compra es: ${total}") 
            self.procesar_pago(total) 

    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente not in self.clientes: 
            print("El cliente no existe")

        else:
              for compra in self.clientes[nombre_cliente]['compras']:
                    print(f"Producto: {compra['productos']},Total: {compra ['total']}")

    def calcular_ventas_totales(self):
        print (f" Las ventas totales han sido de ${self.ventas_totales}")

    def realizar_compra (self):
        carrito = {}
        self.ver_inventario()
        while True:
            compra = input ("¿Qué producto quieres hoy?:")
            if compra == "salir":
                print ("Hasta Pronto")
                break

            for elemento in self.inventario:
                if elemento['nombre'] == compra: 
                    try:
                        unidades_pedidas = int(input(f"¿Cuántas unidades de {compra} quieres? (Stock disponible: {elemento['cantidad']})"))
                        if unidades_pedidas > elemento['cantidad']:        
                            print(f"No hay suficiente stock")
                        else:                                             
                            carrito[compra] = {'precio': elemento['precio'], 'cantidad': unidades_pedidas}
                            elemento['cantidad'] -= unidades_pedidas
                            break
                    except:                                               
                        print("Carácter incorrecto, introduce un número")
                        
                    break
            else:                                  
                print(f"El producto {compra} no existe en el inventario")
        return carrito

    def procesar_pago(self, total_compra):
        try:
            pago = float (input ("Importe entregado:"))
            if pago >= total_compra:
                cambio = float (pago - total_compra)
                print (f"Operación realizada con éxito, su cambio es de {cambio}")

            else:
                print (f"Importe insuficiente, le faltan ${total_compra - pago} euros")

        except:
            print ("Caracter incorrecto, por favor introduzca un importe válido")

```
